```python
# convert datatype to numpy array
# astype() function
import numpy as np
a = np.array([1,2,3,4,5])
print("original array:",a)
print("original datatype:",a.dtype)

# convert datatype to float
new_arr = a.astype('float')
print("converted array:",new_arr)
print("converted datatype:",new_arr.dtype)
```

```python
# Handling conversion errors
import numpy as np
a = np.array(['1','2','3','4','5'])
print("original array:",a)
print("original datatype:",a.dtype)
# convert datatype to int
new_arr = a.astype('int')
print("converted array:",new_arr)
print("converted datatype:",new_arr.dtype)

def convert(arr,target_type):
    try:
        new_arr = arr.astype(target_type)
        print("converted array:",new_arr)
        print("converted datatype:",new_arr.dtype)
    except ValueError:
        print("Conversion not possible")
d_type = convert(a,'int8')
if d_type is not None:
    print("Data type after conversion:",d_type)
    print("Conversion successful")
else:
    print("Conversion failed")
```

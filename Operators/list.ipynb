```python
#list comprehension
matrix=[[1,2,3],[4,5,6],[7,8,9]]
odd_number = [element for row in matrix for element in row if element %2!=0]
print(odd_number)
```

```python
def find_second_highest(numbers):
    largest=max(numbers)
    numbers.remove(largest)
    return max(numbers)
numbers=[10,22,33,43,21,90]
second_highest=find_second_highest(numbers)
print(second_highest)
```

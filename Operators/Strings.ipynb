```python
#reduce() function
from functools import reduce
num =[1,2,3,4,5]
num2 =[1,2,3,4]
result =  reduce(lambda x,y: x+y, num)
result2 = reduce(lambda x,y:x*y,num2)

print(result)
print(result2)
```

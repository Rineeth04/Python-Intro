```python
#sets are unordered collections of unique elements
# they are mutable
# they are defined using curly braces or the set() constructor
a= set()
b= {1,2,3,4,5}
print(type(a))
print(b)
```

```python
#dictionaries are unordered collections of key-value pairs
# they are mutable
a = dict()
b={"key1": "value1", "key2": "value2", "key3": "value3"}
print(type(a))
print(b)
print(b["key1"])
print(b.get("key2"))
```

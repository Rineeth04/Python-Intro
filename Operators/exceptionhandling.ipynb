```python
try:
    a=int(input("enter the value of a:"))
    b=int(input("enter the value of b:"))
    c=a/b
    print("The answer is divided by b:",c)
except ValueError:
    print("entered value is wrong..")
except ZeroDivisionError:
    print("can't divide the value..")
```

```python
def simple_interest(amount,year,rate):
    try:
        if rate>100:
            raise ValueError(rate)
        interest=(amount*year*rate)/100
        print("The simple interest:",interest)
        return interest
    except ValueError:
        print("interest out of range--",rate)
print('case1:')
simple_interest(5000,1,10)
print('case2:')
simple_interest(10000,3,22)
print('case3:')
simple_interest(2000,1,101)
```

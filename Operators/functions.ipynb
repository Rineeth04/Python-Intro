```python
principle =0
rate = 0
time = 0

while principle<=0:
    principle = int(input("enter the principle amount: "))
    if principle<=0:
        print("principle rate cannot be less than o or negative..")
while rate<=0:
    rate = int(input("enter the rate of interest : "))
    if rate<=0:
        print("interest rate cannot be less than o or negative..")
while time<=0:
    time = int(input("enter the time in years: "))
    if time<=0:
        print("principle rate cannot be less than o or negative..")
total = principle * pow((1+rate/100),time)
print(f"balance after {time} years: Rs:{total:.2f}")
```

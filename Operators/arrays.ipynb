```python
#searching elements in array
import array as arr
a = arr.array('i',[1, 2, 3, 1, 2, 5])
print("The new created array is:",end=" ")
for i in range(0,6):
    print(a[i],end=" ")
print("\r")
print("The index of ")
print(a.index(1))
print(a.index(5))
```

```python
# count elements in array
import array as arr
a = arr.array('i',[1,2,3,3,2,1,4,4])
print(a.count(2))
print(a.count(3))
```

```python
import array as arr
a = arr.array('i',[1,2,4,1,2,5])
print('the  created array is :' ,end=' ')
for i in range(0,6):
    print(a[i],end=' ')
print('\r')
a.reverse()
print('The new array is:',end=' ')
for i in range(len(a)):
    print(a[i],end=' ')
```

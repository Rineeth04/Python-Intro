```python
# Access Modifiers in Python
# In Python, there are three types of access modifiers: public, protected, and private.
# 1. Public Access Modifier:
#    - Public members are accessible from anywhere in the program.
#    - They can be accessed both inside and outside the class.

# 2. Protected Access Modifier:
#    - Protected members are accessible within the class and its subclasses.
#    - They are denoted by a single underscore prefix (e.g., _protected_member

# 3. Private Access Modifier:
#    - Private members are accessible only within the class where they are defined.
#    - They are denoted by a double underscore prefix (e.g., __private_member).
```

```python
#public access modifier
class MyClass:
    def __init__(self):
        self.public_variable = "This is a public variable"
my_object = MyClass()
print(my_object.public_variable)  # Output: This is a public variable
```

```python
#Protected access modifier
class vehicle:
    def __init__(self,brand,model,country):
        self._brand = brand
        self._model = model
        self._country = country
class car(vehicle):
    def show(self):
        print("brand: ", self._brand, "model: ", self._model, "country: ", self._country)
c1 = car("BMW", "X5", "Germany")
print(c1._brand)  # Output: BMW
c1.show()  # Output: brand:  BMW model:  X5 country:
```

```python
#Private access modifier
class person:
    def __init__(self,name,age):
        self.__name = name
        self.__age = age
    def display(self):
        print("Name: ", self.__name, "Age: ", self.__age)
p1 = person("Rineeth", 24)
print(p1.__name)  # This will raise an AttributeError
p1.display()  # Output: Name:  Rineeth Age:  24
```

```python
# Merging Dataframes
import pandas as pd
df1 = pd.DataFrame({'id1':[1,2,3,4],'name':['Rineeth','kiran','Sunny','Trisha']})

df2 = pd.DataFrame({'id2':[3,4,5,6],'course':['Python','ML','Java','AI']})

inner_result = pd.merge(df1,df2, how='inner' , on='id')
inner_result
```

```python
outer_result = pd.merge(df1,df2, how='outer' , on='id')
outer_result
```

```python
left_result = pd.merge(df1,df2, how='left' , on='id')
left_result
```

```python
right_result = pd.merge(df1,df2, how='right' , on='id')
right_result
```

```python
result = pd.concat([df1,df2],axis=0, ignore_index=True)
result
```

```python
result = pd.concat([df1,df2],axis=1)
result
```

```python
result = pd.concat([df1,df2],axis=0, keys=['df1','df2'])
result
```

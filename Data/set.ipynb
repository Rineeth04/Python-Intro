```python
# sets are unordered collection of unique elements. 
# They are mutable, meaning you can add or remove elements from a set after it has been created.
#  Sets are defined using curly braces {},
# Sets are uniques we can't able to insert duplicate values in the set.
s = {1, 2, 3, 4, 5}
print(type(s))
print("Length of set s:", len(s))

```

```python
numbers = {10, 20, 30, 40, 50}
print("Maximum value in set numbers:", max(numbers))
print("Minimum value in set numbers:", min(numbers))
print("Sum of all values in set numbers:", sum(numbers))
```

```python
set1 = {1, 2, 3, 4, 5}
for num in set1:
    print(num)
```

```python
chars = set("python")   # {'p','y','t','h','o','n'}
print(chars)
```

```python
# remove and discard
set1 ={1, 2, 3, 4, 5}
set1.remove(3)  # removes 3 from the set
print("Set after removing 3:", set1)
set1.discard(4)  # removes 4 from the set
print("Set after discarding 4:", set1)
set1.discard(10)  # does nothing, since 10 is not in the set
print("Set after discarding 10:", set1)
```

```python
animals = frozenset(["cat", "dog", "rabbit"])
print("Frozenset of animals:", animals)
```

```python
# set comprehension
nums =[2,4,5,7]
even = {num for num in nums if num % 2 == 0}
print("Even numbers from the list:", even)
```

```python
# check if an element is in the set
my_set = {1, 2, 3, 4, 5}
x=3
print(x in my_set)  # True
```

```python
# Add an element to the set
set1 ={1,2,3}
set2={4,5,6}
set1.update(set2)  # set1 now contains {1, 2, 3, 4, 5, 6}
print("Set after update:", set1)
```

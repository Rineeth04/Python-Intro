```python
t = (1, 2, 3, 4, 5)
print(type(t))
print("Length of tuple t:", len(t))
print("Maximum value in tuple t:", max(t))
print("Minimum value in tuple t:", min(t))
print("Sum of all values in tuple t:", sum(t))
print("Average of all values in tuple t:", sum(t)/len(t))
```

```python
# packing and unpacking of tuple
t1 = 10, 20, 30
print("Tuple t1:", t1)
a, b, c = t1
print("Values after unpacking t1:")
print("a:", a)
print("b:", b)
print("c:", c)
```

```python
# Iterating the tuple
t1= (1,8,"Rineeth", 4.5)
for tuple in t1:
    print(tuple)
```

```python
# find within the range
tuple1 = (10, 20, 30, 40, 50, 60, 70, 80)
#index out of range
position= tuple1 .index(10)
print(position)
```

```python
# count the occurrences of a value in the list
tuple1 = (10, 20, 30, 40, 50, 60, 70, 80,20,)
count = tuple1.count(20)
print("Count of value 20:", count)
```

```python
# Nested Tuples
nested_tuple = ((1, 2), (3, 4), ("Python", "Java"))
print("Nested Tuple:", nested_tuple)
print(nested_tuple[2][0])

for i in nested_tuple:
    for j in i:
        print(j, end=' ')

    print("\n")
```

```python
text = "apple banana apple"
freq = {}
for word in text.split():
    freq[word] = freq.get(word, 0) + 1
print(freq)
```

```python
students = {"Rineeth": [85, 90, 88],"venepally": [78, 80, 79],"Rao": [92, 95, 94]
}
for name, marks in students.items():
    avg = sum(marks) / len(marks)
    print(name, "Average:", avg)
```

```python
numbers = [1,2,3,4,5,6]

grouped = {"even": [], "odd": []}

for num in numbers:
    if num % 2 == 0:
        grouped["even"].append(num)
    else:
        grouped["odd"].append(num)

print(grouped)
```

```python
# Nested Dictionary
data ={"user1": {"age": 25, "city": "Hyderabad"},
    "user2": {"age": 30, "city": "Mumbai"}}
for user, details in data.items():
        print(user,details["city"])
        print(user,details["age"])
```

```python
metrics = {
    "accuracy": 0.91,
    "precision": 0.88,
    "recall": 0.85
}

for metric, value in metrics.items():
    print(metric, ":", value)
```

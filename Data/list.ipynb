```python
# Accessing the list of elements
my_list = [1,2,3,4,5]
print(my_list[0]) # Accessing the first element
print(my_list[-1])

# slicing the list
print(my_list[1:4]) # Accessing elements from index 1 to 3
print(my_list[:3]) # Accessing the first three elements

#Accessing the elements which are not present in the list
# This will raise an IndexError
try:
    print(my_list[10])
except IndexError:
    print("Index out of range")
```

```python
# get the list of items in the list using the for loop
list = ["Rineeth",22,True,"Rao",[1,2,3]]
for item in list:
    print(item)
```

```python
#Iterate along with an index number
my_list= [10,12,"Rineeth","Rao",8.1]
for i in range(0, len(my_list)): # range(0,5)
    print(my_list[i])
   # print(f"Index: {i}, Value: {my_list[i]}")
```

```python
# Adding elements to the list using append() method
list1 = [1,2,3]
list1.append("Add")
list1.append([2,"Rao",10])
print(list1)

# Add item in specific index using insert() method
list1.insert(1,"Rineeth")
print("After adding item to list1:", list1)

# extend the list using extend() method
list2=["Rineeth","Rao",10]
list1.extend(list2)
print("After extending list1 with list2:", list1)
```

```python
# Modifying the list
list1= [1,2,3,"Rineeth",10]
list1[3] = "Rao"
print("After modifying list1:", list1)

#modify by using the slice assignment
list1[1:3] = ["Rineeth", "Rao"]
print("After modifying list1 using slice assignment:", list1)
```

```python
# Modify all items in the list
my_list = [2, 4, 6, 8]

# change value of all items
for i in range(len(my_list)): # range(0,4)
    # calculate square of each number
    square = my_list[i] * my_list[i]
    my_list[i] = square

print(my_list)
```

```python
# remove the elements from the list using remove() method
list1 = [1,2,3,"Rineeth",10]
list1.remove(3)
print("After removing item from list1:", list1)

my_list =[6, 4, 6, 6, "Rineeth8", 12]

for item in my_list:
    my_list.remove(6)

print(my_list)

my_list.pop()
print(my_list)

# clear the list using clear() method
my_list.clear()
print(my_list)
```

```python
# Find the index of an item in the list using index() method
list1 = [1,2,3,"Rineeth",10]
index = list1.index("Rineeth")
print("Index of 'Rineeth' in list1:", index)

#Extends() method is used to add all the items of one list to another list. It takes a list as an argument and adds all the items of that list to the end of the current list. The extend() method modifies the original list and does not return a new list.
list1 =["Toyota", "Honda", "Ford"]
list2 = ["BMW", "Audi"]
list1.extend(list2)
print("After extending list1 with list2:", list1)

#copy() method is used to create a shallow copy of a list. It returns a new list that is a copy of the original list. The copy() method does not modify the original list and can be used to create a new list with the same elements as the original list.
list1 = ["Toyota", "Honda", "Ford"]
list2 = list1.copy()
print("Original list1:", list1)
print("Copied list2:", list2)
```

```python
# sort() method in list
list1=["Rineeth","java","Python"]
list1.sort()
print("Sorted list1:", list1)

# reverse() method in list
list1.reverse()
print("Reversed list1:", list1)

```

```python
# library functions for list
l1 = [1, 2, 3, 4, 5]
print("Length of l1:", len(l1))
print("Maximum value in l1:", max(l1))
print("Minimum value in l1:", min(l1))
print("Sum of all values in l1:", sum(l1))
print("Average of all values in l1:", sum(l1)/len(l1))
print(all(l1))
print(any(l1))
```

```python
# Nested list 
l= [[1, 2, 3], ["Rineeth", "Rao"], [4.5, 6.7]]
for sublist in l:
    for item in sublist:
        print(item)
```

```python
# List comprehension
a=[1,2,3,4]
result = [val*2 for val in a]
print(result)

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
squared_matrix = [[val**2 for val in row] for row in matrix]
print(squared_matrix)
```

```python
# creating the even numbers list using list comprehension
even_numbers = [num for num in range(1, 12) if num % 2 == 0]
print(even_numbers)
```

```python
# defining a class in python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    # behavior of the person(class)
    def display(self):
        print("name: ", self.name)
        print("age: ", self.age)
p1 = person("John", 30)
p1.display()
```

```python
class student:
    school_name = "ABC School"  # class variable
    def __init__(self, name, age):
        self.name = name  # instance variable
        self.age = age    # instance variable
s1 = student("Alice", 20)

# accessing instance variables
print("Student Name: ", s1.name)
print("Student Age: ", s1.age)

# accessing class variable
print("School Name: ", student.school_name)

# Modify instance variables
s1.name="Rineeth"
s1.age=23
print("Student Name: ", s1.name)
print("Student Age: ", s1.age)

#Modify class variable
student.school_name = "XYZ School"
print("School Name: ", student.school_name)
```

```python
#Modify the object properties
class student:
    def __init__(self,name,grade,school):
        self.name = name
        self.grade = grade
        self.school = school
    def display(self):
        print("Name:", self.name, "Grade:", self.grade, "School:", self.school)

# create an instance using the new class definition
s1 = student("Alice", "A", "ABC School")

# Modify the object properties
s1.name = "Rineeth"
s1.grade = "A+"

s1.display()

#del s1.school
#print(s1.school)  # This will raise an AttributeError since 'school' has been deleted
```

```python
#Default constructor with parameters
class person:
    def __init__(self,name="Rineeth",age=24):
        self.name = name
        self.age = age
p1 =person()
print("Name: ", p1.name)
print("Age: ", p1.age)
```

```python
#Non-parameterized constructor
class company:
    def __init__(self):
        self.company_name = "ABC Company"
        # assign the location string directly
        self.location = "Hyderabad"
c1 = company()
print("Company Name: ", c1.company_name)
print("Location: ", c1.location)
```
